### Day 9: Recursion

###### Recursive Method for Calculating Factorial


##### Function Description 
Complete the factorial function in the editor below. Be sure to use recursion.<br>
*factorial has the following paramter:*<br>
* int n: an integer
Returns
* int: the factorial of n <br>

`Note:` If you fail to use recursion or fail to name your recursive function factorial or Factorial, you will get a score of 0. <br>
`Input Format`<br>
A single integer, *n* (the argument to pass to factorial).<br>
**Constraints**

* Your submission must contain a recursive function named factorial.

```python
def factorial(n):
    if (n < 0):
        return -1
    elif (n == 0):
        return 1
    else:
        return n * factorial(n - 1)
x = int(input("Enter the number: "))
print(f"Factorial of {x} is:",factorial(x))
```
